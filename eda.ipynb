```python
user_profile = pd.json_normalize(users.profile_details)
user_profile.head()
```

```python
users = pd.concat([users, user_profile], axis=1)
```

```python
users = users.drop(columns=["profile_details", "course_links", "undefined_description_title"])
users = users[users.user_id != 0]
```

```python
users.rename(columns={"user_id": "id"}, inplace=True)
users.rename(columns={"teacher_name": "name"}, inplace=True)
```

```python
users.email.fillna("", inplace=True)
```

```python
users["is_teacher"] = users.role == "Cán bộ"
```

```python
users.loc[users.role == "", "role"] = None
```

```python
users.head()
```

### normalize

```python
users['name'] = users['name'].str.lower()
users['role'] = users['role'].str.lower()
users['email'] = users['email'].str.lower()
users['country'] = users['country'].str.lower()
users['city'] = users['city'].str.lower()
```

## EDA Course

```python
courses.rename(columns={"course_id": "id"}, inplace=True)
courses.rename(columns={"course_name": "name"}, inplace=True)
courses.rename(columns={"teachers_text": "teachers"}, inplace=True)
courses.drop(columns=["teacher_links"], inplace=True)
courses.head()
```

```python
courses.info()
```

```python
courses.drop_duplicates(subset='id', keep='first', inplace=True)
```

```python
# Extract using regex
courses['course_code'] = courses['name'].str.extract(r'\(([A-Z]{2}\d{4,6})\)').fillna('')
courses['semester'] = courses['name'].str.extract(r'(HK\d{3,4})').fillna('')
courses['program_code'] = courses['name'].str.extract(r'\(([A-Z]{2,4})_HK').fillna('')
courses['program'] = courses['name'].str.extract(r'\[([^\]]+)\]')[0].fillna('')
```

```python
courses.program.value_counts()
```

### normalize

```python
# Normalize text data to lowercase
courses['name'] = courses['name'].str.lower()
courses['teachers'] = courses['teachers'].str.lower()
courses['program'] = courses['program'].str.lower()
courses['program_code'] = courses['program_code'].str.lower()
courses['course_code'] = courses['course_code'].str.lower()
courses['semester'] = courses['semester'].str.lower()
```

```python
courses.head()
```

## EDA course time

```python
courses_time.head()
```

```python
slot = pd.json_normalize(courses_time.subjectDetails)
slot.head()
```

```python
courses_slot = courses_time.explode('subjectDetails').reset_index(drop=True)
courses_slot = pd.concat([courses_slot, pd.json_normalize(courses_slot.subjectDetails)], axis=1)
courses_slot.head()
```

```python
courses_slot.drop(columns=["subjectDetails"], inplace=True)
```

```python
courses_slot["code"] = courses_slot["code"].str.lower()
courses_slot["name"] = courses_slot["name"].str.lower()
courses_slot["teacher"] = courses_slot["teacher"].str.lower()
courses_slot["group"] = courses_slot["group"].str.lower()
courses_slot.rename(columns={"group": "program"}, inplace=True)

courses_slot["language"] = courses_slot["language"].str.lower()
courses_slot["dayOfWeek"] = courses_slot["dayOfWeek"].str.lower()
courses_slot["room"] = courses_slot["room"].str.lower()
courses_slot.head()
```

## EDA users_courses

```python
users_courses.info()
```

```python

invalid_users_courses = pd.concat([
    users_courses[users_courses.course_id == "None"],
    users_courses[users_courses.user_id == 0]
])

print("invalid users courses", invalid_users_courses.shape)

users_courses.drop(invalid_users_courses.index, inplace=True)
```

```python
# Cast course_id to int to match with courses.id
users_courses['course_id'] = users_courses['course_id'].astype(int)
users_courses['user_id'] = users_courses['user_id'].astype(int)
```

## EDA current user/course/slot

```python
current_courses = courses[courses.semester == "hk252"]
current_courses.to_csv("hk252/courses_hk252.csv", index=False)
```

```python
courses_slot.to_csv("hk252/slot_hk252.csv", index=False)
```

```python
users_courses.head()
users_courses.info()
```

```python
# Join current_courses.id with users_courses.course_id
current_user_course = current_courses.merge(
    users_courses, 
    left_on='id',           # Column from current_courses
    right_on='course_id',   # Column from users_courses
    how='inner'             # Join type
)

current_user_course = current_user_course.merge(
    users, 
    left_on='user_id', 
    right_on='id', 
    how='inner',
    suffixes=('_course', '_user')  # Adds suffix to duplicate columns
)
current_user_course.head()
```

```python
current_user_course.drop(columns=["course_id", "id_user"], inplace=True)
current_user_course.rename(columns={"name_course": "course_name", "teachers":"teacher_name", "id_course": "course_id", "name_user":"user_name"}, inplace=True)
```

```python
current_user_course.to_csv("hk252/user_course_hk252.csv", index=False)
```
