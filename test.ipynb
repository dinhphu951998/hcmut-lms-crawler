```python
import os
```

```python
# Get all user IDs from the users folder
users_folder = "users"
user_ids = []

# Check if the folder exists
if os.path.exists(users_folder):
    # Get all files in the users folder
    for filename in os.listdir(users_folder):
        # Check if the file has .html extension
        if filename.endswith(".html"):
            # Extract the ID (remove .html extension)
            user_id = filename.replace(".html", "")
            user_ids.append(int(user_id))
    
    # Sort the user IDs (optional, for better organization)
    user_ids.sort()
    
    # Save to userId.txt file
    with open("userId.txt", "w", encoding="utf-8") as f:
        for user_id in user_ids:
            f.write(str(user_id) + "\n")
    
    print(f"Extracted {len(user_ids)} user IDs and saved to userId.txt")
else:
    print(f"Folder '{users_folder}' does not exist")

```

```python
import json

users = json.load(open("all_users.json", "r", encoding="utf-8"))
links = set(link for user in users for link in user.get("course_links", []) if link and "id=None" not in link)
open("course_link.txt", "w", encoding="utf-8").writelines(f"{link}\n" for link in sorted(links))
print(f"Saved {len(links)} distinct course links")
```
